```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
weather_data = pd.read_csv('/content/NAPANE_MAY2023-JULY2023_update.csv')
weather_data.head(5)
```

```python
weather_data['Precipitation'] = pd.to_numeric(weather_data['Precipitation'], errors='coerce')
```

```python
mean_temperature= weather_data['Temperature'].mean()
mean_Humidity= weather_data['Humidity'].mean()
mean_Precipitation = weather_data['Precipitation'].mean()
print("Mean Temperature :", mean_temperature)
print("Mean Humidity :", mean_Humidity)
print("Mean Precipitation :", mean_Precipitation)
```

```python
data_df_num=weather_data.loc[:,['Temperature','Min_Temperature','Max_Temperature','Humidity','Precipitation']]
data_df_num.head()
```

```python
X = data_df_num
y = data_df_num.pop("Humidity")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
y_pred = regression_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
np.mean(np.absolute(y_pred-y_test))
```

```python
for i in range(len(y_pred)):
  y_pred[i]=round(y_pred[i])
pd.DataFrame({'Actual':y_test,'Prediction':y_pred,'diff':(y_test-y_pred)})
```
