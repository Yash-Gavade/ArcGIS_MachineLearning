```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('/content/NAPANE_MAY2023-JULY2023_update.csv')
data.head()
```

```python
data.shape
```

```python
data.drop(['Surface_Pressure'], axis = 1, inplace = True)
data.head()
```

```python
# missing columns
data.isna().sum()
```

```python
data.describe()
```

**Humidity** (Create a pivot table to aggregate humidity by date & Heatmap Using Seaborn)

```python
humidity_table = pd.pivot_table(data, values='Humidity', index=['Days'])
sns.heatmap(data = humidity_table, cmap='Reds')
```

Temperature

```python

Temperature_table = pd.pivot_table(data, values='Temperature', index=['Days'])
sns.heatmap(data = Temperature_table, cmap='Oranges')
```

Precipitation

```python
Precipitation_table = pd.pivot_table(data, values='Precipitation', index=['Days'])
sns.heatmap(data = Precipitation_table, cmap='Blues')
```
